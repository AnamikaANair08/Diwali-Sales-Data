# data cleaning

```python
#To remove blank columns 

df.drop(['Status', 'unnamed1'], axis=1, inplace = True)
```

```python
#using sum() along with isnull to give the total null values if present

pd.isnull(df).sum()
```

```python
df.shape
```

```python
df.dropna(inplace = True)
```

```python
pd.isnull(df).sum()
```

```python
#To change data type from float to integer

df['Amount']= df['Amount'].astype('int')
```

# Exploratory data analysis

# Gender

```python
df.columns
```

```python
ax = sns.countplot(x='Gender', data =df)

#To mention data labels on the columns 
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by ='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount',data =sales_gen)
```

```python
# from the above graph we can conclude that the female buyers are more than male buyers and even purchasing power of females
# are greater than men.
```

# Age

```python
df.columns
```

```python
ax = sns.countplot(data = df, x='Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Total Amount vs Age Group

sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Age Group', y='Amount', data=sales_age)
```

```python
#For above graph we can see that most of the buyers are of the age group 26-35 years female.
```

# state

```python
df.columns
```

```python
# total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize': (17,5)})
sns.barplot(data=sales_state, x='State', y='Orders')
```

```python
# total sales from top 10 states

amount_states = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc= {'figure.figsize':(17,5)})
sns.barplot(data=amount_states, x='State', y='Amount')
```

```python
# From the above graph we can say that most of the orders and amount are from Uttar pradesh, Maharshtra
# and Karnataka respectively.
```

# Marital Status

```python
ax = sns.countplot(data =df, x='Marital_Status')

sns.set(rc={'figure.figsize':(6,4)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(data= sales_state, x='Marital_Status', y='Amount', hue='Gender')
```

```python
# From the above graph we can see that most of the buyers are married(women) and they have high purchasing power.
```

# Occupation

```python
sns.set(rc={'figure.figsize':(15,5)})
ax = sns.countplot(data=df, x='Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_data = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(data=sales_data, x='Occupation', y='Amount')
```

```python
# from above graph we can see that most of the buyers are working in IT, healthcare and aviation sector
```

```python
df.columns
```

# Product category

```python
sns.set(rc={'figure.figsize':(24,5)})

ax = sns.countplot(data=df, x='Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_data = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,9)})
sns.barplot(data=sales_data, x='Product_Category', y='Amount')
```

```python
# From above graph we can say that most sold products are food, clothing and electronics
```

```python
df.columns
```

```python
sales_data = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_data, x='Product_ID', y='Orders')
```

```python
# Top 10 most sold products (same as above)

fig1, ax1= plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
